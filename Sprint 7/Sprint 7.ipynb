```python
''' *****   Nivell 1 *****
   
   Exercici 1 - Calculadora de l'índex de massa corporal

Escriu una funció que calculi l'IMC ingressat per l'usuari/ària, és a dir, qui ho executi haurà d'ingressar aquestes dades. Pots obtenir més informació del seu càlcul en:

-> Índice de masa corporal IMC que es y como se calcula.

La funció ha de classificar el resultat en les seves respectives categories
'''


def calculoIMC(peso,altura):
    altura=altura/100
    imc=peso /(altura**2)

    if imc < 18.5:
        cat_imc= "Bajo Peso"
    elif imc>=18.5 and imc<24.9:
        cat_imc= "Peso normal"
    elif imc>=25 and imc<29.9:
        cat_imc= "Sobrepeso"
    elif imc>=30:
        cat_imc= "Obesidad"

    resultado=  str(round(imc,1)) + " - " + cat_imc

    return resultado

#peso= input("Introduce peso en kilos")
#altura= input ( "Introduce la altura en centimetros")
#print(calculoIMC(int(peso),int(#altura)))
print(calculoIMC(55,158))
```

```python
'''  Exercici 2 - Convertidor de temperatures.

Existeixen diverses unitats de temperatura utilitzades en diferents contextos i regions. 
Les més comunes són Celsius (°C), Fahrenheit (°F) i Kelvin (K). 
També existeixen altres unitats com Rankine (°Ra) i Réaumur (°Re). Selecciona almenys 2 i construeix el convertidor.

'''

def ConvertidorTemp(valor,unidad_origen, unidad_destino):
   
    resultado=""
    
    # Comprobar los parametros
    if not isinstance(valor, (int, float)): 
        resultado ="El valor ha de ser numérico"
    if not (unidad_origen=="C" or unidad_origen=="F" or unidad_origen=="K"):
        resultado=resultado + " Unidad origen no válida"
    if not (unidad_destino=="C" or unidad_destino=="F" or unidad_destino=="K"):
        resultado=resultado + " Unidad destino no válida"
    
    if resultado=="":
   
        if unidad_origen=="C":
            if unidad_destino=="F":
                resultado= (valor * 9/5)+ 32
            elif unidad_destino=="K":
                resultado= valor + 273.15
        elif unidad_origen=="F":
            if unidad_destino=="C":
                resultado= (valor - 32) * 5 / 9
            elif unidad_destino=="K":
                resultado= (valor - 32) * 5 / 9 + 273.15
        elif unidad_origen=="K":
            if unidad_destino=="C":
                    resultado =  valor - 273.15
            elif unidad_destino=="F":
                resultado = (valor  * 9/ 5) - 459.67  

        resultado = str( round(resultado,3)) + " " + unidad_destino

    return resultado

ConvertidorTemp(50,"C", "F")
```

```python
'''        *****   Nivell 2 *****

Exercici 1 - Diccionari invers amb duplicats

Continuant amb l'exercici 4 del nivell 1: al client es va oblidar de comentar un detall i resulta que els valors en el diccionari original
poden duplicar-se i més, per la qual cosa les claus intercanviades poden tenir duplicats. 
En aquest cas, en l'exercici anterior imprimies un missatge d'advertiment, ara, els valors del diccionari resultant hauran d'emmagatzemar-se
com una llista.
Tingues en compte que si és un valor únic no ha de ser una llista.
'''

def invertirDiccConDuplicados(oldDic):

    dicOldKeys=oldDic.keys()
    dicOldValues=oldDic.values()
    hayRepetidos=False
    newDic=dict.fromkeys(dicOldValues)

    for i in dicOldKeys:  

        if list(oldDic.values()).count(oldDic[i])>1: 
          # hacer lista
            if (newDic[oldDic[i]]==None):
                valores=[]
                for clave, valor in oldDic.items():
                    if valor==oldDic[i]:
                        valores.append(clave)    
                newDic[oldDic[i]]=valores
        else: 
            newDic[oldDic[i]]=i

    return newDic

dicEncuesta={"Preg1":"Resp1","Preg2":"Resp1","Preg3":"Resp3","Preg4":"Resp4","Preg5":"Resp3","Preg6":"Resp6","Preg7":"Resp1"} 
invertirDiccConDuplicados(dicEncuesta)
```

```python
''' Exercici 2 - Conversió de tipus de dades

El client rep una llista de dades i necessita generar dues llistes, la primera on estaran tots els elements que es van poder convertir 
en flotants i l'altra on estan els elements que no es van poder convertir. 

Exemple de la llista que rep el client: [ '1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']]

''' 

def conversorTiposDatos(datos):
    esfloat=[]
    noEsfloat=[]

    for item in datos:
            if (type(item) == str):
                try:   
                    esfloat.append(item)
                except:
                    noEsfloat.append(item)
            elif (type(item)== list or type(item)==tuple):
                for elem in item:
                     try:   
                        float(elem)
                        esfloat.append(elem)
                     except:
                         noEsfloat.append(elem)
    return esfloat,noEsfloat
conversorTiposDatos([ '1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']])
```

```python
''' ************** Nivell 3 **************  

Exercici 1 - Comptador i endreçador de paraules d'un text.

El client va quedar content amb el comptador de paraules, però ara vol llegir arxius TXT i que calculi la freqüència de cada paraula ordenades 
dins de les entrades habituals del diccionari segons la lletra amb la qual comencen, és a dir, 
les claus han d'anar de la A a la Z i dins de la A hem d'anar de la A la Z. 

'''

import re

def ContaryOrdenarPalabras(fichero):
    palabras=[]
    palabrasDef=[]
    palabrasRep=[]
    letrasRep=[]
    clavesLetras=[]
    dicLetras={}
    palabrasRep=[]

    fichero =open(fichero,encoding='utf-8')

    for line in fichero:
        palabras+=line.split()

    for palabra in palabras:
        palabra=palabra.lower()
        palabra = re.sub(r'[^\w]', '', palabra)
        palabrasDef.append (palabra)

    palabrasDef.sort()

    for word in palabrasDef:
        if word!="" and word[0] not in letrasRep:
            clavesLetras.append(word[0])
            letrasRep.append(word[0])
            
    palabrasDef.remove("")
    letrasRep.clear()

    for letra in clavesLetras:
    
        if letra not in letrasRep:
            letrasRep.append(letra)
            dicLetras[letra]=""
            dicPalabras={}
        for pal in palabrasDef: 
            if pal[0]==letra:
                if pal not in palabrasRep:
                    palabrasRep.append(pal)
                    dicPalabras[pal]=palabrasDef.count(pal)
                    dicLetras[letra]=dicPalabras
    return dicLetras

ContaryOrdenarPalabras('C:/Users/Public/tu_me_quieres_blanca.txt')
```
